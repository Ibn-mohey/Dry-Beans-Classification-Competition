```python
train_path = r"../input/dry-beans-classification-iti-ai-pro-intake01/train.csv"
test_path = r"../input/dry-beans-classification-iti-ai-pro-intake01/test.csv"
df = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

```

```python
df["Focus"]=np.sqrt((df["MajorAxisLength"]/2)**2-(df["MinorAxisLength"]/2)**2)
df["focal"]=((df["MajorAxisLength"]/2)**2)/df["Focus"]

df_test["Focus"]=np.sqrt((df_test["MajorAxisLength"]/2)**2-(df_test["MinorAxisLength"]/2)**2)
df_test["focal"]=((df_test["MajorAxisLength"]/2)**2)/df_test["Focus"]
```

```python
df["CompactnessBinned"]=pd.qcut(df["Compactness"], q=2)
df_test["CompactnessBinned"]=pd.qcut(df_test['Compactness'], q=2)

labelencoder = LabelEncoder()

df["CompactnessBinned"] = labelencoder.fit_transform(df["CompactnessBinned"])
df_test["CompactnessBinned"] = labelencoder.fit_transform(df_test["CompactnessBinned"])
```

```python
df["ExtentBinned"]=pd.qcut(df["Extent"], q=13)
df_test["ExtentBinned"]=pd.qcut(df_test['Extent'], q=13)

labelencoder = LabelEncoder()

df["ExtentBinned"] = labelencoder.fit_transform(df["ExtentBinned"])
df_test["ExtentBinned"] = labelencoder.fit_transform(df_test["ExtentBinned"])
```

```python
ranges = [0, 323.747994, np.inf] 
labels = [0,1]
df["MinorAxisLength_labelled"] = pd.cut(df["MinorAxisLength"], 
                                  bins=ranges, 
                                  labels=labels)

df_test["MinorAxisLength_labelled"] = pd.cut(df_test["MinorAxisLength"], 
                                  bins=ranges, 
                                  labels=labels)
ranges = [0, 0.003960, np.inf] 
labels = [1,0]
df["ShapeFactor1_labelled"] = pd.cut(df['ShapeFactor1'], 
                                  bins=ranges, 
                                  labels=labels)

df_test["ShapeFactor1_labelled"] = pd.cut(df_test['ShapeFactor1'], 
                                  bins=ranges, 
                                  labels=labels)
```

```python
df["AreaBinned"]=pd.qcut(df["Area"], q=31)
df_test["AreaBinned"]=pd.qcut(df_test['Area'], q=31)

labelencoder = LabelEncoder()

df["AreaBinned"] = labelencoder.fit_transform(df["AreaBinned"])
df_test["AreaBinned"] = labelencoder.fit_transform(df_test["AreaBinned"])

```

```python
df["CompactnessBinned2"]=pd.qcut(df["Compactness"], q=35)
df_test["CompactnessBinned2"]=pd.qcut(df_test["Compactness"], q=35)

labelencoder = LabelEncoder()

df["CompactnessBinned2"] = labelencoder.fit_transform(df["CompactnessBinned2"])
df_test["CompactnessBinned2"] = labelencoder.fit_transform(df_test["CompactnessBinned2"])

```

```python
x=sorted(df.groupby(by="y").describe()["Solidity"]["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
df["Binned1"] = pd.cut(df["Solidity"], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned1"] = pd.cut(df_test["Solidity"], bins=ranges_SF1, 
                                    labels=labels_SF1)
        
x=sorted(df.groupby(by="y").describe()["roundness"]["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
        
df["Binned2"] = pd.cut(df["roundness"], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned2"] = pd.cut(df_test["roundness"], bins=ranges_SF1, 
                                    labels=labels_SF1)
        
x=sorted(df.groupby(by="y").describe()["Compactness"]["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
        
df["Binned3"] = pd.cut(df["Compactness"], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned3"] = pd.cut(df_test["Compactness"], bins=ranges_SF1, 
                                    labels=labels_SF1)
        
x=sorted(df.groupby(by="y").describe()['ShapeFactor1']["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
        
df["Binned4"] = pd.cut(df['ShapeFactor1'], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned4"] = pd.cut(df_test['ShapeFactor1'], bins=ranges_SF1, 
                                    labels=labels_SF1)
        
        
        
        
x=sorted(df.groupby(by="y").describe()['ShapeFactor2']["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
df["Binned5"] = pd.cut(df['ShapeFactor2'], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned5"] = pd.cut(df_test['ShapeFactor2'], bins=ranges_SF1, 
                                    labels=labels_SF1)
        
x=sorted(df.groupby(by="y").describe()['ShapeFactor3']["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
        
df["Binned6"] = pd.cut(df['ShapeFactor3'], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned6"] = pd.cut(df_test['ShapeFactor3'], bins=ranges_SF1, 
                                    labels=labels_SF1)
        
x=sorted(df.groupby(by="y").describe()['ShapeFactor4']["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
        
df["Binned7"] = pd.cut(df['ShapeFactor4'], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned7"] = pd.cut(df_test['ShapeFactor4'], bins=ranges_SF1, 
                                    labels=labels_SF1)
        
x=sorted(df.groupby(by="y").describe()["Focus"]["50%"].tolist())
ranges_SF1= x+[np.inf]
ranges_SF1.insert(0,0)
labels_SF1 = [0,1,2,3,4,5,6,7]
        
df["Binned8"] = pd.cut(df["Focus"], bins=ranges_SF1, 
                                    labels=labels_SF1)
df_test["Binned8"] = pd.cut(df_test["Focus"], bins=ranges_SF1, 
                                    labels=labels_SF1)
```

```python
selected_features=['Area', 'Perimeter', 'MajorAxisLength', 'MinorAxisLength'
                       ,'AspectRation','Eccentricity', 'ConvexArea','EquivDiameter'
                       , 'Extent','Solidity', 'roundness', 'Compactness',"ShapeFactor4","Focus"
                           ,"Binned1","Binned2","Binned3","Binned4","Binned5","Binned6","Binned7","Binned8"
                           ,"AreaBinned"
                           ,"ExtentBinned","ShapeFactor1_labelled","MinorAxisLength_labelled","CompactnessBinned"]
X1 = df[selected_features]
y1 = df["y"]

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.2, stratify=y1,random_state=42,shuffle=True,)
clf = CatBoostClassifier(cat_features=["Binned1","Binned2","Binned3","Binned4","Binned5","Binned6","Binned7","Binned8","AreaBinned","ExtentBinned","ShapeFactor1_labelled","MinorAxisLength_labelled","CompactnessBinned"])
clf.fit(X_train, y_train)
y_pred_test=clf.predict(X_test)


accuracy = f1_score(y_test, y_pred_test, average = 'macro')
print(accuracy)
```

```python
y_pred_test_final=clf.predict(df_test[selected_features])
df_submission=pd.DataFrame()
df_submission["ID"]=df_test["ID"]
df_submission["y"]=y_pred_test_final
df_submission[["ID", "y"]].to_csv('./8Binned.csv', index=False)
```
